O AUC (Area Under the Curve - Área Abaixo da curva) representa o grau ou medida de separabilidade. 

Quanto maior o AUC, melhor o modelo está em prever 0s como 0s e 1s como 1s. 

ROC (Receiver Operating Characteristics - Característica de Operação do Receptor) é uma curva de probabilidade. 

É criada calculando a taxa verdadeiro-positivo contra a taxa de falsos-positivos. Ou seja, o número de vezes que o classificador acertou a predição conta o número de vezes que o classificador errou a predição.

Como plotar uma curva AUC?

```python
from sklearn import metrics                           # Metricas para calcular accuracy score
from sklearn.linear_model import LogisticRegression   # Modelo utilizado
from sklearn.model_selection import train_test_split  # Separa dados de treinamento e teste
from sklearn.datasets import load_breast_cancer       # Carrega o dataset Breast Cancer    
import matplotlib.pyplot as plt                       # Plotagem de gráficos
```

O próximo passo é separar os dados do dataset em dados de teste e dados de treinamento e criar um modelo. 

À princípio será utilizada a técnica técnica de "regressão logística" utilizando a biblioteca "sklearn". 

Scikit-learn é uma biblioteca de aprendizado de máquina de código aberto que oferece suporte ao aprendizado supervisionado e não supervisionado. Também fornece várias ferramentas para ajuste de modelo, pré-processamento de dados, seleção e avaliação de modelo e muitos outros utilitários. 

Para construção do primeiro modelo, será utilizada a implementação LogisticRegression da bibioteca skitlearn.
A regressão logística, apesar do nome, é um modelo linear para classificação em vez de regressão. 
A regressão logística também é conhecida na literatura como regressão logit, classificação de entropia máxima (MaxEnt) ou classificador log-linear.

A regressão logística implementada em LogisticRegression permite ajustar a regressão logística binária, One-vs-Rest ou multinomial com regularização opcional ou Elastic-Net. 

Os dados utilizados neste notebook para plotar a curva AUC são oriundos do dataset Breast Cancer, encontrados em https://archive.ics.uci.edu/ml/datasets.php. 

```python
# Carregando Breast Cancer Dataset
breast_cancer = load_breast_cancer()
X = breast_cancer.data
y = breast_cancer.target

# Separando o Dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=44)

# Criando um modelo 
clf = LogisticRegression(penalty='l2', C=0.1)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
# Accuracy
print("Accuracy", metrics.accuracy_score(y_test, y_pred))

#AUC Curve
y_pred_probability = clf.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_probability)
auc = metrics.roc_auc_score(y_test, y_pred_probability)
plt.plot(fpr,tpr,label="LogisticRegression, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

Comparando LogisticRegression e um modelo de Árvore de Decisão Simples. 

O novo código cria um novo modelo baseado em árvore e plota a AUC da árvore de decisão junto com a AUC do modelo anterior.

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

y_pred_probability = clf.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_probability)
auc = metrics.roc_auc_score(y_test, y_pred_probability)
plt.plot(fpr,tpr,label="DecisionTree, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

O modelo com DecisionTree possui um desempenho inferior ao do LogisticRegression, pois a área sob a curva é menor no segundo modelo.

Outro modelo bastante utilizado em classificação é o RandomForest.

O código para adicionar o RandomForest é semelhante ao DesicionTree.

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

y_pred_probability = clf.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_probability)
auc = metrics.roc_auc_score(y_test, y_pred_probability)
plt.plot(fpr,tpr,label="RandomForest, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

O desempenho do LogistRegression ainda é superior que os demais.
